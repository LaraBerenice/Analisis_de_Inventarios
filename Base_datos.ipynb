```python
#autenticacion SQLserver
import pyodbc
import os
from dotenv import load_dotenv

# Cargar las variables de entorno desde el archivo .env
load_dotenv()

# Obtener las variables de entorno
hostname = os.getenv("DB_HOST")
db = os.getenv("DB_NAME")
user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")

# Cadena de conexión usando usuario y contraseña
connection = pyodbc.connect(f'DRIVER={{SQL Server}};SERVER={hostname};DATABASE={db};UID={user};PWD={password}')

# Verificar la conexión
print("Conexión exitosa a la base de datos")

# Crear un cursor para ejecutar consultas
cursor = connection.cursor()
```

```python
# Diccionario con los nombres de los archivos
archivos = {
    'precios_compra_dic': r"archive\2017PurchasePricesDec.csv",
    'Inicio_Inv_FINAL': r"archive\BegInvFINAL12312016.csv",
    'final_inv_Final': r"archive\EndInvFINAL12312016.csv",
    'Factura_Compras': r"archive\InvoicePurchases12312016.csv",
    'Compras_Finales': r"archive\PurchasesFINAL12312016.csv",
    'Ventas_Final': r"archive\SalesFINAL12312016.csv"
}

# Leer y mostrar las primeras 5 filas de cada archivo
for nombre, ruta in archivos.items():
    ruta_archivo = os.path.join(directorio_actual, ruta)
    # Verificar si el archivo existe antes de intentar leerlo
    if os.path.exists(ruta_archivo):
        df = pd.read_csv(ruta_archivo)
        print(f"\n{nombre}:\n", df)
    else:
        print(f"El archivo {ruta_archivo} no existe.")
```

```python
# Leer, contar filas y limpiar nulos de cada archivo
for nombre, ruta in archivos.items():
    ruta_archivo = os.path.join(directorio_actual, ruta)
    
    # Verificar si el archivo existe antes de intentar leerlo
    if os.path.exists(ruta_archivo):
        # Leer el archivo CSV
        df = pd.read_csv(ruta_archivo)
        
        # Contar filas antes de eliminar nulos
        filas_originales = df.shape[0]
        
        # Eliminar filas con nulos
        df_limpio = df.dropna()
        
        # Contar filas después de eliminar nulos
        filas_limpias = df_limpio.shape[0]
        
        # Imprimir resultados
        print(f"\n{nombre}:")
        print(f"Total de filas originales: {filas_originales}")
        print(f"Total de filas después de eliminar nulos: {filas_limpias}")
        
    else:
        print(f"El archivo {ruta_archivo} no existe.")
```

```python
# Función para crear tabla si no existe
def create_table_if_not_exists(create_table_sql):
    # Ejecutar la consulta para verificar si la tabla existe
    table_name = create_table_sql.split()[2]  # Obtener el nombre de la tabla
    check_table_sql = f'''
    IF NOT EXISTS (SELECT * FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_NAME = '{table_name}')
    BEGIN
        {create_table_sql}
    END
    '''
    cursor.execute(check_table_sql)

# Definir las consultas SQL para crear las tablas
create_precios_compra_dic = '''
CREATE TABLE precios_compra_dic (
    Brand INT,
    Description VARCHAR(255),
    Price FLOAT,
    Size VARCHAR(255),
    volume INT,
    Classification INT,
    PurchasePrice FLOAT,
    VendorNumber INT,
    VendorName VARCHAR(255)
);
'''

create_inicio_inv_final = '''
CREATE TABLE Inicio_Inv_FINAL (
    InventoryId VARCHAR(255),
    Store INT,
    City VARCHAR(255),
    Brand INT,
    Description VARCHAR(255),
    Size VARCHAR(255),
    onHand INT,
    Price FLOAT,
    startDate DATE
);
'''

create_final_inv_final = '''
CREATE TABLE final_inv_Final (
    InventoryId VARCHAR(255),
    Store INT,
    City VARCHAR(255),
    Brand INT,
    Description VARCHAR(255),
    Size VARCHAR(255),
    onHand INT,
    Price FLOAT,
    endDate DATE
);
'''

create_factura_compras = '''
CREATE TABLE Factura_Compras (
    VendorNumber INT,
    VendorName VARCHAR(255),
    InvoiceDate DATE,
    PONumber INT,
    PODate DATE,
    PayDate DATE,
    Quantity INT,
    Dollars FLOAT,
    Freight FLOAT,
    Approval FLOAT
);
'''

create_compras_finales = '''
CREATE TABLE Compras_Finales (
    InventoryId VARCHAR(255),
    Store INT,
    Brand INT,
    Description VARCHAR(255),
    Size VARCHAR(255),
    VendorNumber INT,
    VendorName VARCHAR(255),
    PONumber INT,
    PODate DATE,
    ReceivingDate DATE,
    InvoiceDate DATE,
    PayDate DATE,
    PurchasePrice FLOAT,
    Quantity INT,
    Dollars FLOAT,
    Classification INT
);
'''

create_ventas_final = '''
CREATE TABLE Ventas_Final (
    InventoryId VARCHAR(255),
    Store INT,
    Brand INT,
    Description VARCHAR(255),
    Size VARCHAR(255),
    SalesQuantity INT,
    SalesDollars FLOAT,
    SalesPrice FLOAT,
    SalesDate DATE,
    Volume INT,
    Classification INT,
    ExciseTax FLOAT,
    VendorNo INT,
    VendorName VARCHAR(255)
);
'''

# Ejecutar las consultas para crear las tablas si no existen
create_table_if_not_exists(create_precios_compra_dic)
create_table_if_not_exists(create_inicio_inv_final)
create_table_if_not_exists(create_final_inv_final)
create_table_if_not_exists(create_factura_compras)
create_table_if_not_exists(create_compras_finales)
create_table_if_not_exists(create_ventas_final)

# Confirmar cambios en la base de datos
connection.commit()

# Cerrar el cursor y la conexión
cursor.close()
connection.close()

print("Tablas creadas exitosamente si no existían.")
```

```python
# Cargar los DataFrames
dataframes = {}
for nombre, archivo in archivos.items():
    dataframes[nombre] = pd.read_csv(archivo)

# Función para manejar las conversiones con errores
def safe_convert(df, column, dtype):
    if dtype == 'int':
        df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0).astype(int)
    elif dtype == 'float':
        df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0.0).astype(float)
    else:
        df[column] = df[column].astype(dtype)

# precios_compra_dic
safe_convert(dataframes['precios_compra_dic'], 'Brand', 'int')
dataframes['precios_compra_dic']['Description'] = dataframes['precios_compra_dic']['Description'].astype(str)
safe_convert(dataframes['precios_compra_dic'], 'Price', 'float')
dataframes['precios_compra_dic']['Size'] = dataframes['precios_compra_dic']['Size'].astype(str)
safe_convert(dataframes['precios_compra_dic'], 'Volume', 'int')
safe_convert(dataframes['precios_compra_dic'], 'Classification', 'int')
safe_convert(dataframes['precios_compra_dic'], 'PurchasePrice', 'float')
safe_convert(dataframes['precios_compra_dic'], 'VendorNumber', 'int')
dataframes['precios_compra_dic']['VendorName'] = dataframes['precios_compra_dic']['VendorName'].astype(str)

# Inicio_Inv_FINAL
dataframes['Inicio_Inv_FINAL']['InventoryId'] = dataframes['Inicio_Inv_FINAL']['InventoryId'].astype(str)
safe_convert(dataframes['Inicio_Inv_FINAL'], 'Store', 'int')
dataframes['Inicio_Inv_FINAL']['City'] = dataframes['Inicio_Inv_FINAL']['City'].astype(str)
safe_convert(dataframes['Inicio_Inv_FINAL'], 'Brand', 'int')
dataframes['Inicio_Inv_FINAL']['Description'] = dataframes['Inicio_Inv_FINAL']['Description'].astype(str)
dataframes['Inicio_Inv_FINAL']['Size'] = dataframes['Inicio_Inv_FINAL']['Size'].astype(str)
safe_convert(dataframes['Inicio_Inv_FINAL'], 'onHand', 'int')
safe_convert(dataframes['Inicio_Inv_FINAL'], 'Price', 'float')
dataframes['Inicio_Inv_FINAL']['startDate'] = pd.to_datetime(dataframes['Inicio_Inv_FINAL']['startDate'], errors='coerce')

# final_inv_Final
dataframes['final_inv_Final']['InventoryId'] = dataframes['final_inv_Final']['InventoryId'].astype(str)
safe_convert(dataframes['final_inv_Final'], 'Store', 'int')
dataframes['final_inv_Final']['City'] = dataframes['final_inv_Final']['City'].astype(str)
safe_convert(dataframes['final_inv_Final'], 'Brand', 'int')
dataframes['final_inv_Final']['Description'] = dataframes['final_inv_Final']['Description'].astype(str)
dataframes['final_inv_Final']['Size'] = dataframes['final_inv_Final']['Size'].astype(str)
safe_convert(dataframes['final_inv_Final'], 'onHand', 'int')
safe_convert(dataframes['final_inv_Final'], 'Price', 'float')
dataframes['final_inv_Final']['endDate'] = pd.to_datetime(dataframes['final_inv_Final']['endDate'], errors='coerce')

# Factura_Compras
safe_convert(dataframes['Factura_Compras'], 'VendorNumber', 'int')
dataframes['Factura_Compras']['VendorName'] = dataframes['Factura_Compras']['VendorName'].astype(str)
dataframes['Factura_Compras']['InvoiceDate'] = pd.to_datetime(dataframes['Factura_Compras']['InvoiceDate'], errors='coerce')
safe_convert(dataframes['Factura_Compras'], 'PONumber', 'int')
dataframes['Factura_Compras']['PODate'] = pd.to_datetime(dataframes['Factura_Compras']['PODate'], errors='coerce')
dataframes['Factura_Compras']['PayDate'] = pd.to_datetime(dataframes['Factura_Compras']['PayDate'], errors='coerce')
safe_convert(dataframes['Factura_Compras'], 'Quantity', 'int')
safe_convert(dataframes['Factura_Compras'], 'Dollars', 'float')
safe_convert(dataframes['Factura_Compras'], 'Freight', 'float')
safe_convert(dataframes['Factura_Compras'], 'Approval', 'float')

# Compras_Finales
dataframes['Compras_Finales']['InventoryId'] = dataframes['Compras_Finales']['InventoryId'].astype(str)
safe_convert(dataframes['Compras_Finales'], 'Store', 'int')
safe_convert(dataframes['Compras_Finales'], 'Brand', 'int')
dataframes['Compras_Finales']['Description'] = dataframes['Compras_Finales']['Description'].astype(str)
dataframes['Compras_Finales']['Size'] = dataframes['Compras_Finales']['Size'].astype(str)
safe_convert(dataframes['Compras_Finales'], 'VendorNumber', 'int')
dataframes['Compras_Finales']['VendorName'] = dataframes['Compras_Finales']['VendorName'].astype(str)
safe_convert(dataframes['Compras_Finales'], 'PONumber', 'int')
dataframes['Compras_Finales']['PODate'] = pd.to_datetime(dataframes['Compras_Finales']['PODate'], errors='coerce')
dataframes['Compras_Finales']['ReceivingDate'] = pd.to_datetime(dataframes['Compras_Finales']['ReceivingDate'], errors='coerce')
dataframes['Compras_Finales']['InvoiceDate'] = pd.to_datetime(dataframes['Compras_Finales']['InvoiceDate'], errors='coerce')
dataframes['Compras_Finales']['PayDate'] = pd.to_datetime(dataframes['Compras_Finales']['PayDate'], errors='coerce')
safe_convert(dataframes['Compras_Finales'], 'PurchasePrice', 'float')
safe_convert(dataframes['Compras_Finales'], 'Quantity', 'int')
safe_convert(dataframes['Compras_Finales'], 'Dollars', 'float')
safe_convert(dataframes['Compras_Finales'], 'Classification', 'int')

# Ventas_Final
dataframes['Ventas_Final']['InventoryId'] = dataframes['Ventas_Final']['InventoryId'].astype(str)
safe_convert(dataframes['Ventas_Final'], 'Store', 'int')
safe_convert(dataframes['Ventas_Final'], 'Brand', 'int')
dataframes['Ventas_Final']['Description'] = dataframes['Ventas_Final']['Description'].astype(str)
dataframes['Ventas_Final']['Size'] = dataframes['Ventas_Final']['Size'].astype(str)
safe_convert(dataframes['Ventas_Final'], 'SalesQuantity', 'int')
safe_convert(dataframes['Ventas_Final'], 'SalesDollars', 'float')
safe_convert(dataframes['Ventas_Final'], 'SalesPrice', 'float')
dataframes['Ventas_Final']['SalesDate'] = pd.to_datetime(dataframes['Ventas_Final']['SalesDate'], errors='coerce')
safe_convert(dataframes['Ventas_Final'], 'Volume', 'int')
safe_convert(dataframes['Ventas_Final'], 'Classification', 'int')
safe_convert(dataframes['Ventas_Final'], 'ExciseTax', 'float')
safe_convert(dataframes['Ventas_Final'], 'VendorNo', 'int')
dataframes['Ventas_Final']['VendorName'] = dataframes['Ventas_Final']['VendorName'].astype(str)
```
